## Тестирование


**Тестирование программного обеспечения** — процесс исследования, испытания программного продукта, имеющий своей целью проверку соответствия  реальныого поведения программы и её ожидаемого поведения. Тестирование проводится на конечном наборе тестов, выбранных определённым образом.  



Когда Вы запускаете на выполнение весь Colab-ноутбук или всё приложение, вводите данные, нажимаете на кнопки, и т.д.  вы, фактически, тестируете работу своего кода. Это - интеграционное тестирование. Современные приложения достаточно сложны и содержат множество зависимостей. Интеграционное тестирование проверяет, что несколько компонентов системы работают вместе правильно. 

Оно выполняет свою задачу, но сложно для автоматизации. Как правило, тесты требуют, чтобы вся или почти вся система была развернута и сконфигурирована на машине, на которой они выполняются. Это ресурсоёмко, сложно, часто приводит к "наведенным" ошибкам.  

Лучше сначала спуститься на предыдущий уровень и убедиться, что компоненты (функции, классы, модули) работают правильно по-отдельности.


**Модульное тестирование** (unit testing, unit-тестирование, юнит-тестирование, автономное тестирование) — процесс в программировании, позволяющий проверить на корректность отдельные модули (функции, классы...) исходного кода программы.

Идея состоит в том, чтобы писать тесты для каждой нетривиальной функции или метода. Это позволяет достаточно быстро проверить, не привело ли очередное изменение кода к появлению ошибок в уже оттестированных местах программы, а также облегчает обнаружение и устранение таких ошибок. 

Подробнее см. статью о [модульном тестировании](https://habr.com/ru/post/169381/). 



Цель модульного тестировани - изолировать каждую часть системы, чтобы убедиться, что они работают, как указано. Использование этого типа тестов на протяжении всей реализации позволяет уменьшить количество ошибок в приложении. Он работает путем сравнения вывода функции, которую необходимо протестировать, с ожидаемыми значениями.

Unit-тесты могут быть написаны собственноручно, без использования сторонних библиотек, а можно использовать специализированные framework’и. На сегодняшний день практически всегда используется второй вариант.

### unittest

**Unittest** – framework для тестирования, входящий в стандартную библиотеку языка Python. 


Рассмотрим простейший модуль Python (`str_functions.py`), который содержит функции для обработки строк, и разберем пример того, как можно было бы его протестировать с использованием `unittest`. 

Чтобы не прописывать полный футь к `str_functions.py`, разместим его в папке по умолчанию - /content

````python
import unittest
# импортируем тестируемые функции
from str_functions import count_lowers, count_digits 
````

Можно было бы разместить все тесты в том же файле, что и сама программа (мы делали это, называя **самотестированием**, и этап самотестирования никто не отменял!), но размещение внешних тестов в отдельном файле  имеет много преимуществ:

- Модуль с тестом может быть запущен автономно из командной строки.
- Тестовый код может быть легко отделён от программы.
- У программиста-разработчика нет возможности изменить тесты для соответствия коду программы без видимой причины.
- Тестовый код должен изменяться гораздо реже, чем программа.
- Протестированный код может быть легче переработан.
- Если стратегия тестирования изменяется, нет необходимости изменения кода программы.

Разберём только одну из концепций автоматизации тестирования, поддерживаемую **unittest** - тестовые случаи (`test case`). 

Базовые блоки тестирования это тестовые случаи - простые случаи, которые должны быть проверены на корректность.

Тестовый случай создаётся путём наследования от класса `unittest`.`TestCase`.

Тестирующий код должен быть самостоятельным и никак не зависеть от других тестов.

Простейший подкласс `TestCase` может просто реализовывать тестовый метод (метод, обязательно начинающийся с `test`). Это важно  для реализации других концепций, когда автоматически собираются и запускаются все методы, описанные во всех классах, наследниках `unittest.TestCase`. 

````python
class TestStrMethods(unittest.TestCase):
# тестовые методы должны начинаться с test  
  def test_count_lowers(self):
      """ Тестирование функций работы со строками"""
      self.assertEqual(count_lowers(''), 0)         
      self.assertEqual(count_lowers('my numbers is 2'), 11)

  def test_count_digits(self):
      """ Тестирование функций работы со строками"""
      self.assertEqual(count_digits(''), 0)
      
````

В данном примере мы использовали единственную функцию `unittest` - 
`assertEqual(a, b)`, которая проверяет равенство своих аргументов. 
Помимо этой функции  модуль `unittest` предоставляет множество других функций для самых различных проверок:

```
assertEqual(a, b) равносильно (далее будем обозначать через "—") a == b

assertNotEqual(a, b) — a != b

assertTrue(x) — bool(x) is True

assertFalse(x) — bool(x) is False

assertIs(a, b) — a is b


```

Полный список функций unittest можно посмотреть  [в статье (на русском языке)](https://pythonworld.ru/moduli/modul-unittest.html) или в ["родной  документации" по unittest ](https://docs.python.org/3/library/unittest.html)

Чтобы запустить написанные тесты, надо обратиться к функции `main()` из `unittest`. Она обеспечивает и интерфес командной строки и запуск из Google Colab, но немного по-разному. Начнем с запуска в Google Colab. В минимальном случае надо вызвать функцию `unittest.main()` с некоторыми аргументами. 

````python
unittest.main(argv=[''], exit=False)
````

Вызов тестирования можно также как в случае самотестирования вставить в условный оператор:

````python
if __name__ == '__main__':
    unittest.main(argv=[''], 
                  defaultTest='TestStrMethods',  # можно явно указать, какие тесты запускать
                  exit=False)
````

Вызов `unittest.main()` под условным оператором особенно важен, если тесты запускаются потом в командной строке. 

В Colab-ноутбуках, можно обойтись и без условного оператора и даже без указания запускаемого по умолчанию теста.  

````python
unittest.main(argv=[''], verbosity=2, exit=False)
````

Несколько слов об аргументах функции `unittest.main()` для запуска в Google Colab.

Почему явно надо указать `argv=['']`? 
Причина в том, что `unittest.main()` просматривает `sys.argv`, а первый параметр — это то, что запустило `IPython` или `Jupyter`. На локальном компьютере проблем нет, а в облаке права и полномочия ваших процессов невелики, поэтому, если явно не указать, что список аргументов пуст, возникнет ошибка о том, что файл подключения к ядру `(/root)` не является допустимым атрибутом. 
Передача **явного пустого списка** в `unittest.main` "обяжет" `IPython` и `Jupyter` НЕ пытаться просматривать` sys.argv`.   

`verbosity=2` - добавление "многословности" - позволяет посмотреть более подробную диагностику при тестировании

`exit=False` предотвратит завершение процесса unittest.main, это позволит нам дальше работать в ноутбуке.

## Code review (код-ревью)

Стадия проверки кода (код-ревью) обычно выполняется кем-то из более опытных программистов, участвующих в работе над проектом. 

Задачи проверки кода: 

*   Поиск ошибок. Многие ошибки могут быть найдены при тестировании, но существуют ошибки, которые не "покрыты" тестами.
*   Анализ стиля — пишете ли вы так, как принято в компании.
*   Правильность оформления кода — соблюдаете ли вы отступы и табуляцию, чтобы код было легче читать. Пишете ли комментарии — если в компании принято комментировать ключевые моменты в коде.
*   Проверка адекватности реализации — вы предложили собственное  изящное (или не очень изящное) решение или использовали библиотечную функцию? А что уместнее в этой ситуации?
*   Проверка степени влияния на проект — если вы полностью переписали формат передачи данных на сервер, то это значит, что другим участникам нужно будет подстроиться под вас и переписать свою часть. **Это дорого**.
*    Анализ уязвимостей в безопасности — может ли что-то пойти не так, если злоумышленник решит использовать этот код в своих целях.

Можно ещё расширить список задач, выполняемых при анализе кода. Если проблемы найдены, проверяющий отправляет код на доработку. Если всё хорошо, код переходит на следующую стадию — как правило, в тестирование.
